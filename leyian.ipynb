```python
import pandas as pd

# Load the Iris dataset
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv")
print(df.head())

print(df.info())  # Shows data types and non-null counts
print(df.isnull().sum())  # Check for missing values

# Drop or fill missing values if any (Iris dataset has none, but here's how you'd handle it)
df = df.dropna()  # OR df.fillna(method='ffill', inplace=True)
print(df.describe())

grouped = df.groupby('species').mean()
print(grouped)



```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Optional for better plot aesthetics
sns.set(style="whitegrid")

# Simulate a trend using row index
df['index'] = df.index
plt.figure(figsize=(10, 5))
sns.lineplot(data=df, x='index', y='sepal_length', hue='species')
plt.title("Sepal Length Trend by Species")
plt.xlabel("Index (simulated time)")
plt.ylabel("Sepal Length")
plt.legend(title="Species")
plt.show()

plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='sepal_width', kde=True)
plt.title("Distribution of Sepal Width")
plt.xlabel("Sepal Width")
plt.show()

plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='sepal_length', y='petal_length', hue='species')
plt.title("Sepal Length vs. Petal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.legend(title="Species")
plt.show()




```
